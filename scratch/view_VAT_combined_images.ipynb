```python
import os
import rasterio
import matplotlib.pyplot as plt

# Set your directory path containing the .tif visualisation files
input_dir = '/Users/jamesbyers/code/github/Kaggle/openai_to_z/scratch/RVT/VAT_images'

# List all .tif files (change filtering if you want e.g. only 'RIB' files)
tif_files = [fn for fn in os.listdir(input_dir) if fn.endswith('.tif')]

print("Found", len(tif_files), "tif files:")

for tif in tif_files:
    print(f"Opening: {tif}")
    file_path = os.path.join(input_dir, tif)
    with rasterio.open(file_path) as src:
        img = src.read(1)  # read the first band
        plt.figure(figsize=(8, 8))
        plt.imshow(img, cmap='gray')
        plt.title(tif)
        plt.colorbar()
        plt.axis('off')
        plt.show()
        # # Optional: Wait for user to continue, or break if you want to stop at any point:
        # nextimg = input("Press Enter for next image, or 'q' to quit: ")
        # if nextimg.lower().startswith('q'):
        #     break
```
